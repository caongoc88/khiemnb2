```python
# Bài 1: (Dễ): Nhập vào một chuỗi gồm các số tự nhiên cách nhau một khoảng trắng.
# Tính tổng, tính trung bình cộng các số. Tìm số âm lớn nhất và số dương bé nhất trong chuỗi vừa nhập.
N = str(input('Nhập chuỗi số tự nhiên:'))
A=0
B=0
max=-10**999
min=10**999
N=N.strip()
N=N.split()
print(N)
for i in range (len(N)):
    A+=int(N[i])
    B+=1
print(f"Tổng của chuỗi = {A}")
print(f"Số thành phần = {B}")
print(f"Trung bình cộng = {A/B}")

for j in range(len(N)):
    if int(N[j]) < 0 and int(N[j])>max:
        max = int(N[j])
    elif int(N[j]) > 0 and int(N[j])< min:
        min = int(N[j])
print(f"Số âm lớn nhất = {max}")
print(f"Số dương bé nhất = {min}")
        
```
