```python
from tkinter import *
from tkinter import ttk
top = Tk()
top.geometry("500x300")
top.title("Thuế TNCN")

def tinhthue(event):
# -Biểu thuế luỹ tiến:
#     +Từ 0-5tr: 5%
#     +Từ 5-10tr:10%
#     +Từ 10-18: 15%
#     +Từ 18-32: 20%
#     +Từ 32-52: 25%
#     +Từ 52-80: 30%
#     +Trên 80: 35%
    a = entry1.get()
    b = entry2.get()
    if a.isnumeric() == False or b.isnumeric() ==False:
        TPN = f'Thông tin không hợp lệ'
    else:
        TNTT = int(a)- 11 - (int(b)*4.4)
        if TNTT <0:
            TPN = f'0.0'
        else:
            if TNTT<=5:
                TPN = TNTT*5/100     
            elif TNTT<=10:
                TPN = (5*5 + (TNTT-5)*10)/100
            elif TNTT<=18:
                TPN = (5*5 + 5*10 + (TNTT-10)*15)/100
            elif TNTT<=32:
                TPN = (5*5 + 5*10 + 8*15 + (TNTT-18)*20)/100
            elif TNTT<=52:
                TPN = (5*5 + 5*10 + 8*15 + 14*20 + (TNTT-32)*25)/100
            elif TNTT<=80:
                TPN = (5*5 + 5*10 + 8*15 + 14*20 + 20*25 + (TNTT-52)*30)/100
            elif TNTT>80:
                TPN = (5*5 + 5*10 + 8*15 + 14*20 + 20*25 + 28*30 + (TNTT-80)*35)/100
    label4['text'] = f'{round(TPN,3)}'

def FocusOut(event):
    a = entry1.get()
    b = entry2.get()
    if a.isnumeric() == False or b.isnumeric() ==False:
        lb_thuesuat2['text'] = 'Thuế suất không xác đinh'
    else:
        TNTT = int(a)- 11 - (int(b)*4.4)
        if TNTT <0:
            lb_thuesuat2['text'] = f'Không phải nộp thuế'
        else:
            if TNTT<=5:
                valuesmax = '5%'     
            elif TNTT<=10:
                valuesmax = '10%'
            elif TNTT<=18:
                valuesmax = '15%'
            elif TNTT<=32:
                valuesmax = '20%'
            elif TNTT<=52:
                valuesmax = '25%'
            elif TNTT<=80:
                valuesmax = '30%'
            elif TNTT>80:
                valuesmax = '35%'
            variable.set(valuesmax)
            lb_thuesuat2['text'] = ''
fontBold = ("Timenewroman","10",'bold')    
label1 = Label(top, text="Tổng thu nhập (trđ)")
label1.place(x=5, y=10)
label2 = Label(top, text="Số người phụ thuộc")
label2.place(x=5, y=60)

entry1 = Entry(top,text='1')
entry1.place(x=130, y=10)
entry2 = Entry(top)
entry2.place(x=130, y=60)
entry2.bind("<FocusOut>",FocusOut)

lb_thuesuat = Label(top,text='Mức thuế suất max')
lb_thuesuat.place(x=5, y=100)
lb_thuesuat2 = Label(top,text='')
lb_thuesuat2.place(x=200, y=100)
values = ('5%','10%','15%','20%','25%','30%','35%')
variable = StringVar()
combobox = ttk.Combobox(top, values = values, state = 'readonly', textvariable = variable)
combobox.place(x=130, y = 100, width = 50)

button = Button(top, text="Tính thuế", font = fontBold)
button.place(x=130, y= 150)
button.bind("<Button-1>", tinhthue)
button.bind("<Return>", tinhthue)

label3 = Label(top, text="Tổng thuế TNCN cần nộp (Trđ):")
label3.place(x= 5, y =200)
label4 = Label(top, text='', fg='red',font =fontBold,width=10)
label4.place(x= 180, y =200)

top.mainloop()
```
